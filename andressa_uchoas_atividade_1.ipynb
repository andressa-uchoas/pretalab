#Projeto 1: Sistema de Gestão de Notas de Alunos

**Lidando com Arquivos, Pacotes e Módulos**

Objetivo: O objetivo deste projeto é desenvolver um sistema de gestão de notas de alunos em Python, que faça uso de manipulação de arquivos CSV, módulos e pacotes.

O sistema deve gerar um arquivo CSV que contenha a média final de cada aluno e o status de aprovado ou reprovado.

**Módulo 1**. Leitura e manipulação de arquivos CSV:

```python
# Importação de bibliotecas
import pandas as pd

# Importação da tabela de dados
base_nota = pd.read_csv('/content/notas.csv', delimiter=',',header=0, )
```

**Módulo 2.1**. Manipulação de arquivos

```python
# Leitura da tabela importada

import sqlite3

# Conectando ao banco de dados (substitua 'meu_banco.db' pelo nome do seu banco)
conn = sqlite3.connect('meu_banco.db')

# Criando um cursor
cursor = conn.cursor()
print(base_nota)
```

```python
base_nota.info()
```

**Módulo 2.2**. Análise de notas

```python
# Importação de biblioteca NumPy para operações condicionais
import numpy as np

# Criando o DataFrame a partir do .csv importado
df= pd.DataFrame(base_nota)

# Calculando a média de cada aluno
df['Média'] = df[['Nota 1', 'Nota 2', 'Nota 3']].mean(axis=1).round(1)
df['Status Final'] = np.where(df['Média'] >= 6,'Aprovação','Reprovação')

df
```


**Módulo 2.3**. Geração de relatórios

```python
# Gerando um arquivo CSV
df.to_csv('resultado_final',index=False)
```
